```python
#Feature Selection by Recursive Backward Elimination
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X = X[cols]
    Xc = sm.add_constant(X)
    model = sm.OLS(y,Xc).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features = cols
print(selected_features)
```

```python
X_new=X[selected_features]
X_new.head()
```

```python
X_new_c=sm.add_constant(X_new)
```

```python
model2=sm.OLS(y,X_new_c).fit()
```

```python
from sklearn.feature_selection import chi2,SelectKBest
```

```python
from sklearn.preprocessing import minmax_scale
Xminmax=minmax_scale(X)
bestfeatures=SelectKBest(score_func=chi2,k=5)
fit=bestfeatures.fit(Xminmax,y)
dfscores=pd.DataFrame(fit.scores_)
dfcolumns=pd.DataFrame(X.columns)
featscore=pd.concat([dfcolumns,dfscores],axis=1)
featscore.columns=['Specs','Score']
top5=featscore.nlargest(5,'Score')
top5
```

```python
sns.barplot(y=top5['Specs'],x=top5['Score'])
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
model.fit(Xminmax,y)
feature_imp=pd.Series(model.feature_importances_, index=X.columns)
feature_imp.nlargest(5).sort_values(ascending=False).plot(kind='bar',color=['c','pink','teal'])
plt.show()
```
